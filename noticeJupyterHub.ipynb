# Notice JupyterHub :

Pour gérer des fichiers sur le serveur JupyterHub du lycée, le plus pratique est d'utiliser la ligne de commande dans un terminal.

Pour ouvrir un terminal cliquez sur le bouton `Nouveau` en haut à droite de la page d'accueil puis choisissez `Terminal`.



Le texte `jupyter-pnom@r610b:~` indique que vous êtes connecté sur la machine nommée `r610b` sous le nom d'utilisateur `jupyter-pnom` et que le dossier de travail actuel est le dossier `~`, c'est-à-dire votre dossier personnel `/home/jupyter-pnom`.

Ce texte `jupyter-pnom@r610b:~$` s'appelle un prompt, une invite de commande, c'est-à-dire une invitation à entrer une commande système... 

## Pour télécharger un fichier :

```python
wget URL_de_téléchargement
```

Exemples :

- Pour télécharger un notebook depuis le nbviewer jupyter, faire un clic droit sur la flèche de téléchargement et choisir `Copier l'adresse du lien` puis la coller dans la ligne de commande telle que par exemple :

```python
wget https://nbviewer.jupyter.org/github/ericECmorlaix/1NSI_2020-2021/blob/master/index.ipynb
```

- Pour télécharger une archive d'un dépot GitHub, faire un clic droit sur `Download ZIP`


et choisir `Copier l'adresse du lien` puis la coller dans la ligne de commande telle que par exemple :


```python
wget https://github.com/ericECmorlaix/1NSI_2020-2021/archive/master.zip
```

Il reste alors à décompresser ce fichier d'archive...

### Dézipper un fichier :

```python
unzip master.zip
```
